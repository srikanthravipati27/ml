```python
#indexing and selecting

d = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'score': [85, 90, 95]
}, index=['a', 'b', 'c'])
print(d['name'])       # Select column  Series
print(d.loc['b'])        # Select row by label  Series
print(d.iloc[1])         # Select row by position  Series
print(d[0:2])            # Slice rows  DataFrame
print(d[d['score'] > 85])  # Boolean index  DataFrame
```

```python
#When you do operations like df1 + df2, pandas aligns both row indices and column names. If a label is not present in both, the result is NaN.
df1 = pd.DataFrame(np.random.randn(3, 3), columns=["A", "B", "C"], index=[0, 1, 2])
df2 = pd.DataFrame(np.random.randn(2, 2), columns=["A", "B"], index=[0, 1])

print(df1 + df2)
```

```python
#When subtracting a Series from a DataFrame, pandas broadcasts row-wise by default  Series index is matched to DataFrame columns.
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6],
    "C": [7, 8, 9]
})

row0 = df.iloc[0]   # Series: A=1, B=4, C=7

df - row0
```

```python
#Arithmetic with scalars like * 5 + 2 is applied element-wise on each value.
df * 2 + 1
df
```
